```python
# ðŸ”¹ Step 1: Import libraries
import nltk
import random

# Download resources (only first time)
nltk.download('punkt')
nltk.download('punkt_tab') # Download punkt_tab as suggested by the error

# ðŸ”¹ Step 2: Define intents (questions + responses)
college_responses = {
    "courses": [
        "We offer B.Tech, M.Tech, MBA, and Diploma programs.",
        "Our college provides undergraduate and postgraduate engineering courses."
    ],
    "admission": [
        "Admissions are based on entrance exams and merit lists.",
        "You can apply online through the college admission portal."
    ],
    "fees": [
        "The fee structure varies by course. For example, B.Tech is around â‚¹75,000 per year.",
        "Please visit the college website for the detailed fee structure."
    ],
    "location": [
        "Our college is located in Hyderabad, Telangana.",
        "The campus is near the city center with excellent transport facilities."
    ],
    "hostel": [
        "Yes, we have separate hostels for boys and girls with good facilities.",
        "Hostel accommodation is available for outstation students."
    ],
    "contact": [
        "You can reach us at info@college.edu or call +91-9876543210.",
        "Please contact the admin office during working hours for details."
    ],
    "default": [
        "I'm sorry, I didn't understand that. Could you please rephrase?",
        "Please ask about courses, admissions, fees, or location."
    ]
}

# ðŸ”¹ Step 3: Simple function to identify keywords
def chatbot_response(user_input):
    user_input = user_input.lower()
    tokens = nltk.word_tokenize(user_input)

    if any(word in tokens for word in ["course", "branch", "program"]):
        return random.choice(college_responses["courses"])
    elif any(word in tokens for word in ["admission", "apply", "join"]):
        return random.choice(college_responses["admission"])
    elif any(word in tokens for word in ["fee", "fees", "payment"]):
        return random.choice(college_responses["fees"])
    elif any(word in tokens for word in ["location", "where", "place"]):
        return random.choice(college_responses["location"])
    elif any(word in tokens for word in ["hostel", "room", "stay"]):
        return random.choice(college_responses["hostel"])
    elif any(word in tokens for word in ["contact", "phone", "email"]):
        return random.choice(college_responses["contact"])
    else:
        return random.choice(college_responses["default"])

# ðŸ”¹ Step 4: Chat loop
print("ðŸ¤– Hello! Iâ€™m your College Enquiry Chatbot. Type 'bye' to exit.\n")

while True:
    user_input = input("You: ")
    if user_input.lower() in ["bye", "exit", "quit"]:
        print("Bot: Thank you! Have a great day ðŸ˜Š")
        break
    print("Bot:", chatbot_response(user_input))
```
